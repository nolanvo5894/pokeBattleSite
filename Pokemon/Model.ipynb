```python
combats = pd.read_csv('combats.csv')
combats.head()
```

```python
sum(combats['second'].isna())
```

```python
detailedCombats = combats
newColumns = ['Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk',
       'Sp. Def', 'Speed', 'Legendary']
#writing stats for first and second movers into the detailedCombats data frame
for num in ['first','second']:
    for col in newColumns:
        detailedCombats[num + col] = poke[col][combats[num]].reset_index()[col]
detailedCombats.info()
```

```python
detailedCombats['Winner'] = (detailedCombats['winner']/detailedCombats['second'] == 1)
detailedCombats['Winner'] = detailedCombats['Winner'].astype('int64')
detailedCombats.head()
```

```python
detailedCombats['firstLegendary'] = detailedCombats['firstLegendary'].astype('int64')
detailedCombats['secondLegendary'] = detailedCombats['secondLegendary'].astype('int64')
detailedCombats.head()
```

```python
# typeChart = pd.read_csv('typeChart.csv')
# typeChart.set_index('Attacking', inplace = True)
# joblib.dump(typeChart,'typeChart.pkl')
```

```python
typeChart = pd.read_csv('typeChart.csv', index_col = 'Attacking')
typeChart
```

```python
first = 'Rock'
second = 'Water'
typeChart[first][second]
```

```python
def getTypeMatch(df):
    firstType1 = df['firstType 1']
    firstType2 = df['firstType 2']
    secondType1 = df['secondType 1']
    secondType2 = df['secondType 2']
    typeMatch = typeChart[secondType1][firstType1] * typeChart[secondType2][firstType2] * typeChart[secondType2][firstType1] * typeChart[secondType1][firstType2]
    return typeMatch
joblib.dump(getTypeMatch,'getTypeMatch.pkl')
```

```python
test = detailedCombats
test['typeMatch'] = test.apply(lambda df: getTypeMatch(df), axis = 1)
test.head()
```

```python
test.columns
```

```python
from sklearn.model_selection import cross_val_score
X = test[['firstHP',
       'firstAttack', 'firstDefense', 'firstSp. Atk', 'firstSp. Def',
       'firstSpeed', 'firstLegendary',
       'secondHP', 'secondAttack', 'secondDefense', 'secondSp. Atk',
       'secondSp. Def', 'secondSpeed', 'secondLegendary', 'typeMatch']]
X = X.iloc[:35000]
y = test['Winner']
y = y.iloc[:35000]
```

```python
from sklearn.ensemble import RandomForestClassifier
pokeForest = RandomForestClassifier()
forestScores = cross_val_score(pokeForest, X, y, cv = 10, scoring = 'accuracy')
forestScores.mean()
```

```python
pokeForest.fit(X,y)
```

```python

joblib.dump(pokeForest, 'pokeModel.pkl')
```
